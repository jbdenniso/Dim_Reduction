```python
# Using ICA to reduce dimensionality creates a shift in the null distrobution and may possibly effect the 'real'
#classification rate despite being used to ID 'functional networks'.
#We're going to write a script that tests and compares multiple 'decomposition' feaure selection usinig the Haxby data
# Factor Analysis, PCA, ICA, dictionary learning
from nilearn import datasets
data=[]
# By default 2nd subject will be fetched
haxby_dataset = datasets.fetch_haxby(n_subjects=6)

# print basic information on the dataset
print('Mask nifti image (3D) is located at: %s' % haxby_dataset.mask)
print('Functional nifti image (4D) is located at: %s' %
      haxby_dataset.func[0])
```

```python

import numpy as np
import matplotlib.pyplot as plt

for sub in [0,1,2,3,4,5]:
    # Load target information as string and give a numerical identifier to each
    behavioral = np.recfromcsv(haxby_dataset.session_target[sub], delimiter=" ")
    conditions = behavioral['labels']

    # Restrict the analysis to faces and places
    condition_mask = np.logical_or.reduce((conditions == b'face', conditions == b'house'))
    tr_con= conditions[condition_mask]
    

    # We now have 2 conditions
    print(np.unique(conditions))
    print tr_con
    session = behavioral[condition_mask]


    from nilearn.input_data import NiftiMasker

    mask_filename = haxby_dataset.mask

    # For decoding, standardizing is often very important
    # note that we are also smoothing the data
    masker = NiftiMasker(mask_img=mask_filename, smoothing_fwhm=4,
                         standardize=True, memory="nilearn_cache", memory_level=1)
    func_filename = haxby_dataset.func[sub]
    X = masker.fit_transform(func_filename)
    # Apply our condition_mask
    X = X[condition_mask]
    
    #Define our Classifier
    from sklearn.svm import SVC
    clf = SVC(kernel='linear',C=1)

    # Define the dimension reduction to be used.
    # Here we use a classical univariate feature selection based on F-test,
    # namely Anova. When doing full-brain analysis, it is better to use
    # SelectPercentile, keeping 5% of voxels
    # (because it is independent of the resolution of the data).
    from sklearn.decomposition import PCA,FastICA,MiniBatchDictionaryLearning,FactorAnalysis
    P=PCA(n_components=20).fit(X).transform(X)
    I=FastICA(n_components=20).fit(X).transform(X)
    DL=MiniBatchDictionaryLearning(n_components=20).fit_transform(X)
    FA=FactorAnalysis(n_components=20).fit_transform(X)
   

     
    from sklearn.model_selection import KFold,permutation_test_score,cross_val_score
    # Define the cross-validation scheme used for validation.
    # Here we use a LeaveOneLabelOut cross-validation on the session label
    # which corresponds to a leave-one-session-out
    cv = KFold(n_splits=6,shuffle=True)
    
    # Compute the prediction accuracy for the different folds (i.e. session)
    # Return the corresponding mean prediction accuracy
    cv_scores = cross_val_score(clf, P, tr_con,cv=6,n_jobs=8)
    classification_accuracy = cv_scores.mean()
    print "**PCA Data**"
    print("Classification accuracy: %.4f / Chance level: %f" %
          (classification_accuracy, 1. / len(np.unique(tr_con))))
    
    cv_scores = cross_val_score(clf, I, tr_con,cv=6)
    classification_accuracy = cv_scores.mean()
    print "**ICA Data**"
    print("Classification accuracy: %.4f / Chance level: %f" %
          (classification_accuracy, 1. / len(np.unique(tr_con))))
    
    cv_scores = cross_val_score(clf, FA, tr_con,cv=6,n_jobs=8)
    classification_accuracy = cv_scores.mean()
    print "**Factor Analysis Data**"
    print("Classification accuracy: %.4f / Chance level: %f" %
          (classification_accuracy, 1. / len(np.unique(tr_con))))
    
    cv_scores = cross_val_score(clf, DL, tr_con,cv=6,n_jobs=8)
    classification_accuracy = cv_scores.mean()
    print "**Dictionary Learning Data**"
    print("Classification accuracy: %.4f / Chance level: %f" %
          (classification_accuracy, 1. / len(np.unique(tr_con))))
    
    
    score,permutation_scores,pvalue=permutation_test_score(clf,P,tr_con,scoring="accuracy",cv=cv,n_permutations=100,n_jobs=8)
    print "**PCA Classification** score %s (pvalue : %s)" %(score,pvalue)
    
    n_classes=2
    plt.hist(permutation_scores,20,label="Permutation scores")
    ylim=plt.ylim()
    
    plt.plot(2*[score],ylim,'--g',linewidth=3,label='Classification Score' '(pvalue %s)'%pvalue)
    plt.plot(2*[0.5],ylim, '--k',linewidth=3,label='Luck')
    plt.ylim(ylim)
    plt.xlabel('Score')
    plt.show()
    
    score,permutation_scores,pvalue=permutation_test_score(clf,I,tr_con,scoring="accuracy",cv=cv,n_permutations=100,n_jobs=8)
    print "**ICA Classification** score %s (pvalue : %s)" %(score,pvalue)
    
    n_classes=2
    plt.hist(permutation_scores,20,label="Permutation scores")
    ylim=plt.ylim()
    
    plt.plot(2*[score],ylim,'--g',linewidth=3,label='Classification Score' '(pvalue %s)'%pvalue)
    plt.plot(2*[0.5],ylim, '--k',linewidth=3,label='Luck')
    plt.ylim(ylim)
    plt.xlabel('Score')
    plt.show()
    
    score,permutation_scores,pvalue=permutation_test_score(clf,FA,tr_con,scoring="accuracy",cv=cv,n_permutations=100,n_jobs=8)
    print "**Factor Analysis Classification** score %s (pvalue : %s)" %(score,pvalue)
    
    n_classes=2
    plt.hist(permutation_scores,20,label="Permutation scores")
    ylim=plt.ylim()
    
    plt.plot(2*[score],ylim,'--g',linewidth=3,label='Classification Score' '(pvalue %s)'%pvalue)
    plt.plot(2*[0.5],ylim, '--k',linewidth=3,label='Luck')
    plt.ylim(ylim)
    plt.xlabel('Score')
    plt.show()
    
    score,permutation_scores,pvalue=permutation_test_score(clf,DL,tr_con,scoring="accuracy",cv=cv,n_permutations=100,n_jobs=8)
    print "**Dictionary Learning Classification** score %s (pvalue : %s)" %(score,pvalue)
    
    n_classes=2
    plt.hist(permutation_scores,20,label="Permutation scores")
    ylim=plt.ylim()
    
    plt.plot(2*[score],ylim,'--g',linewidth=3,label='Classification Score' '(pvalue %s)'%pvalue)
    plt.plot(2*[0.5],ylim, '--k',linewidth=3,label='Luck')
    plt.ylim(ylim)
    plt.xlabel('Score')
    plt.show()
    
```
